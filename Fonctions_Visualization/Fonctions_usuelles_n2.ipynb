## Fonctions usuelles

```python
# Constante
x = np.linspace(-10, 10, 400)

no_graph = np.arange(1, 10, 0.1)

for i, graph in enumerate(no_graph):
    y = np.full_like(x, i)
    plt.plot(x, y)
plt.tight_layout
plt.show()
```
