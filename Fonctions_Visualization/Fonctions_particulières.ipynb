```python
# Importer les librairies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from math import *
import seaborn as sns
from matplotlib import colormaps as cm
```

```python
# cos(x-1)
x = np.linspace(-10, 10, 400)
no_graph = np.arange(1, 10, 1)
for i, graph in enumerate(no_graph):
    plt.plot(x, np.cos(x - i))
    plt.title(f"cos(x- n)")

plt.tight_layout()
plt.show()
```

```python
# cos(x-1)
x = np.linspace(-10, 10, 400)
no_graph = np.arange(1, 10, 1)
for i, graph in enumerate(no_graph):
    plt.plot(x, np.sin(x - i))
    plt.title(f"sin(x- n)")

plt.tight_layout()
plt.show()
```

```python

x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z1 = np.sin(X**2 + Y**2)

plt.contourf(X, Y, Z1, cmap='viridis')
plt.title('sin(x^2 + y^2)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Cos(xy)

Z2 = np.cos(X * Y)
plt.contourf(X, Y, Z2, cmap='plasma')
plt.title('cos(xy)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# (x^2 + y^2)^2 = a^2 * (x^2 - y^2)
x = np.linspace(-10, 10, 5)
y = np.linspace(-10, 10, 5)
X, Y = np.meshgrid(x, y)
Z = (X ** 2 + Y**2)**2

plt.contourf(X, Y, Z)
plt.show()
```

```python
# D√©finir les grilles de x et y
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# D√©finir la constante a
a = 1

# Calculer les valeurs de Z pour l'√©quation (x^2 + y^2)^2 - 2a^2(x^2 - y^2)
Z = (X**2 + Y**2)**2 - 2 * a**2 * (X**2 - Y**2)

# Cr√©er une figure et une sous-graphe
fig, ax = plt.subplots(figsize=(8, 6))

# Tracer la courbe de niveau correspondant √† Z = 0
contour = ax.contour(X, Y, Z, levels=[0], colors='blue')
ax.set_title('Lemniscate de Bernoulli $(x^2 + y^2)^2 = 2a^2(x^2 - y^2)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal', 'box')

plt.show()
```
